```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(palette="pastel")
```

```python
work_dir = "D:/OneDrive - Otto-Friedrich-Universit√§t Bamberg/Master/Masterarbeit/Data/Twitter/"
```

```python
topic_0 = pd.read_csv(work_dir + "topic_0-coded.csv")
topic_1 = pd.read_csv(work_dir + "topic_1-coded.csv")
topic_2 = pd.read_csv(work_dir + "topic_2-coded.csv")

topic_0.code = topic_0.code.str.title()
topic_1.code = topic_1.code.str.title()
topic_2.code = topic_2.code.str.title()
```

```python
topic_0_code = topic_0.groupby(['code']).size().sort_values(ascending=False)
topic_1_code = topic_1.groupby(['code']).size().sort_values(ascending=False)
topic_2_code = topic_2.groupby(['code']).size().sort_values(ascending=False)

plt.figure(figsize=(15,15))
plt.pie(topic_0_code, labels=topic_0_code.index, autopct='%.0f%%',wedgeprops = {'linewidth': 1})
plt.show()

plt.figure(figsize=(15,15))
plt.pie(topic_1_code, labels=topic_1_code.index, autopct='%.0f%%',wedgeprops = {'linewidth': 1})
plt.show()

plt.figure(figsize=(15,15))
plt.pie(topic_2_code, labels=topic_2_code.index, autopct='%.0f%%',wedgeprops = {'linewidth': 1})
plt.show()
```
